```python
#Libraries import
import pandas as pd
import numpy as np

#Plays import
hamlet = pd.read_csv('data/hamlet.tsv', sep = '\t')
macbeth = pd.read_csv('data/macbeth.tsv', sep = '\t')
othello = pd.read_csv('data/othello.tsv', sep = '\t')
```

```python
#Calculating compound mean for the character and "Act - Scene" - Hamlet
mean_hamlet_section = hamlet.groupby(['Act - Scene']).mean()
mean_hamlet_character = hamlet.groupby(['Act - Scene', 'Character']).mean()

#Calculating compound mean for the character and "Act - Scene" - Othello
mean_othello_section = othello.groupby(['Act - Scene', 'compound']).mean()
mean_othello_character = othello.groupby(['Act - Scene', 'Character']).mean()

#Calculating compound mean for the character and "Act - Scene" - Macbeth
mean_macbeth_section = macbeth.groupby(['Act - Scene', 'compound']).mean()
mean_macbeth_character = macbeth.groupby(['Act - Scene', 'Character']).mean()

#Print mean compound mean for the character and "Act - Scene"
print("Hamlet Section", mean_hamlet_section)
print("Hamlet Character", mean_hamlet_character)
print("Macbeth Section", mean_macbeth_section)
print("Macbeth Character", mean_macbeth_character)
print("Othello Section", mean_othello_section)
print("Othello Character", mean_othello_character)

#Export compound mean for the character and "Act - Scene"
mean_hamlet_character.to_csv('mean_hamlet_character.tsv', sep = '\t')
mean_hamlet_section.to_csv('mean_hamlet_section.tsv', sep = '\t')
mean_othello_character.to_csv('mean_othello_character.tsv', sep = '\t')
mean_othello_section.to_csv('mean_othello_section.tsv', sep = '\t')
mean_macbeth_character.to_csv('mean_macbeth_character.tsv', sep = '\t')
mean_macbeth_section.to_csv('mean_macbeth_section.tsv', sep = '\t')
```
